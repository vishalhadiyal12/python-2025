WAP to perform (e.g., "ababccabcc" → "a3b3c4").

```python
s = "ababccabcc"
counts = {}

for i in s:
    if i in counts:
        counts[i] += 1
    else:
        counts[i] = 1

output = ""
for j, count in counts.items():
    output += j + str(count)

print(output)
```

WAP to check whether a number is an Armstrong number.

```python
num = int(input("Enter a number: "))
sum = 0
n = len(str(num))  # number of digits

for i in str(num):
    sum += int(i) ** n

if sum == num:
    print(num, "is an Armstrong number")
else:
    print(num, "is not an Armstrong number")
```

WAP to check whether a number is a Strong number (sum of factorial of digits equals the number).

```python
num = int(input("Enter a number: "))
sum = 0
temp = num

while temp > 0:
    digit = temp % 10
    fact = 1
    
    for i in range(1, digit + 1):
        fact = fact * i
        
    sum = sum + fact
    temp = temp // 10

if sum == num:
    print("Strong Number")
else:
    print("Not Strong Number")
```

WAP to check whether a number is a Perfect number (sum of divisors equals the number).

```python
num = int(input("Enter a number: "))

sum_div = 0

for i in range(1, num):
    if num % i == 0:
        sum_div += i

if sum_div == num:
    print("It is a Perfect number")
else:
    print("It is not a Perfect number")
```

WAP to convert lowercase letters to uppercase without using upper() method.

```python
string = input("Enter a string: ")

result = ""

for ch in string:
    if 'a' <= ch <= 'z':          # check lowercase
        result += chr(ord(ch) - 32)   # convert to uppercase
    else:
        result += ch

print("Uppercase String:", result)
```
